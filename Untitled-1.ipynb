```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df_copy = df.copy()
```

```python
# Define a lambda function to extract the day of the week, month of the year, and year
get_day_month_year = lambda x: (
    x.strftime('%A'),  # Day of the week
    x.strftime('%B'),  # Month of the year
    x.year             # Year
)
# Apply the lambda function to the datetime column
df_copy[['day_of_week', 'month_of_year', 'year']] = df_copy['InvoiceDate'].apply(get_day_month_year).tolist()

```

```python
df_copy.head()
```

```python
# creating a function to classify the time of the day a purchase took place
def classify_time(hour):
    if hour < 12:
        return 'Morning'
    elif hour < 18:
        return 'Afternoon'
    else:
        return 'Night'

# Apply the classify_time function to the hour component of the datetime, and store the result in a new column
df_copy['time_of_day'] = df_copy['InvoiceDate'].dt.hour.apply(classify_time)
```

```python
#Creating a function that finds the total sales a categorical variable and plots a barchart in descending order
def total_sales_by(data, groupby_col, sumby_col):
    total_sales_by_group = data.groupby(groupby_col)[sumby_col].sum().sort_values(ascending=False)
    plt.bar(total_sales_by_group.index, total_sales_by_group.values)
    plt.xticks(rotation=90)
    plt.xlabel(groupby_col.capitalize())
    plt.ylabel(sumby_col.capitalize())
    plt.title('Total Sales by {}'.format(groupby_col.capitalize()))
    plt.show()
```

```python
#Creating a function that finds the total quantity a categorical variable and plots a barchart in descending order
def total_quantity_by(data, quantityby_col, sumby_col):
    total_quantity_by = data.groupby(quantityby_col)[sumby_col].sum().sort_values(ascending=False)
    plt.bar(total_quantity_by.index, total_quantity_by.values)
    plt.xticks(rotation=90)
    plt.xlabel(quantityby_col.capitalize())
    plt.ylabel(sumby_col.capitalize())
    plt.title('Total Quantity by {}'.format(quantityby_col.capitalize()))
    plt.show()
```
